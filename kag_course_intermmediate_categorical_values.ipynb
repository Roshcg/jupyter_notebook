```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = pd.read_csv('/home/m1049296/Downloads/train.csv', index_col='Id')
```

```python
X_test = pd.read_csv('/home/m1049296/Downloads/test.csv', index_col='Id')
```

```python
X.dropna(axis=0, subset=['SalePrice'], inplace=True)
```

```python
y = X.SalePrice
```

```python
X.drop('SalePrice', axis=1, inplace=True)
```

```python
miss_cols = [col for col in X.columns if X[col].isnull().any()]
```

```python
X.drop(miss_cols, axis=1, inplace=True)
```

```python
X_test.drop(miss_cols, axis=1, inplace=True)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=1)
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor()
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

```python
drop_X_train = X_train.select_dtypes(exclude=['object'])
drop_X_valid = X_valid.select_dtypes(exclude=['object'])
```

```python
print('without drop')
print(score_dataset(drop_X_train, drop_X_valid, y_train, y_valid))
```

```python
low_cardinality_cols = [col for col in X_train.columns if X_train[col].nunique() < 10]
```

```python
from sklearn.preprocessing import OneHotEncoder

# Use as many lines of code as you need!
oh_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)

cat_col_train = pd.DataFrame(oh_encoder.fit_transform(X_train[low_cardinality_cols]))
cat_col_valid = pd.DataFrame(oh_encoder.transform(X_valid[low_cardinality_cols]))

cat_col_train.index = X_train.index
cat_col_valid.index = X_valid.index

num_X_train = X_train.drop(low_cardinality_cols, axis=1)
num_X_valid = X_valid.drop(low_cardinality_cols, axis=1)

OH_X_train = pd.concat([num_X_train, cat_col_train], axis=1)
OH_X_valid = pd.concat([num_X_valid, cat_col_valid], axis=1)
```
