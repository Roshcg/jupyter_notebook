```python
import ast 
```

```python
ted['rating_list'] = ted.ratings.apply(ast.literal_eval)
```

```python
pd.options.display.max_colwidth = 50
```

```python
ted.head()
```

```python
def num_ratings(l):
    total = 0
    for d in l:
        total += d['count']
    return total
```

```python
ted['num_rating'] = ted.rating_list.apply(num_ratings)
```

```python
ted.head()
```

```python
ted.rating_list[1]
```

```python
def funny_count(l):
    for d in l:
        if d['id'] == 7:
            return d['count']
```

```python
ted['f_count'] = ted.rating_list.apply(funny_count)
```

```python
ted.head()
```

```python
ted.groupby('speaker_occupation').f_count.sum().sort_values()
```
